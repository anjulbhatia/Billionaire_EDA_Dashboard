### 1. Demographic Analysis of Billionaires 

### 1.1. Distribution of Wealth by Age

```python
# Distribution of Wealth by Age
plt.figure(figsize=(8, 3))
palette = sns.color_palette("Blues_d")
ax = sns.histplot(data=df, x='age', bins=20, kde=True, palette=palette)
for i in ax.containers:
    ax.bar_label(i)
plt.title('Distribution of Wealth by Age Among Billionaires')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Details of the youngest, oldest and median bilionaires
youngest_billionaire = df[df['age'] == df['age'].min()]
oldest_billionaire = df[df['age'] == df['age'].max()]
median_billionaire = df[df['age'] == df['age'].median()]

print("\nDetails of the Youngest Billionaire:")
display(youngest_billionaire[['personName', 'age', 'finalWorth', 'country', 'source']])

print("\nDetails of the Oldest Billionaire:")
display(oldest_billionaire[['personName', 'age', 'finalWorth', 'country', 'source']])

print("\nDetails of the Median Billionaire:")
display(median_billionaire[['personName', 'age', 'finalWorth', 'country', 'source']])
```

***The distribution of wealth by age is a bell shaped curve. The youngest billionaire in the list is Clemente Del Vecchio (age:18) and the oldest is George Joseph (age:101). The median age among billionaires is 65 years.***

### 1.2. Distribution of Wealth by Gender

```python
# Distribution of Wealth by Gender (Area Plot with KDE)
plt.figure(figsize=(8, 4))
sns.kdeplot(data=df, x='finalWorth', hue='gender', fill=True)
plt.title('Distribution of Wealth by Gender Among Billionaires')
plt.xlabel('Final Worth')
plt.ylabel('Density')
plt.legend(title='Gender')
plt.show()
```

```python
# Set up the figure with subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Count of Billionaires by Gender (Bar Plot)
sns.countplot(data=df, x='gender', palette='Blues_d', ax=ax1)
ax1.set_title('Count of Billionaires by Gender', weight='bold',fontsize=12,pad=5)
ax1.set_xlabel('Gender')
ax1.set_ylabel('Count')

# Add labels to the bars
for i, value in enumerate(df['gender'].value_counts().values):
    ax1.text(i, value, f'{value}', ha='center', va='bottom', color='black')

# Distribution of Wealth by Gender (Box Plot)
sns.boxplot(data=df, x='gender', y='finalWorth', palette='Blues_d', ax=ax2)
ax2.set_title('Distribution of Wealth by Gender Among Billionaires', weight='bold',fontsize=12,pad=5)
ax2.set_xlabel('Gender')
ax2.set_ylabel('Final Worth')

# Add median and maximum labels
for i, gender in enumerate(df['gender'].unique()):
    median_value = df[df['gender'] == gender]['finalWorth'].median()
    max_value = df[df['gender'] == gender]['finalWorth'].max()
    ax2.text(i, median_value, f'Median: {int(median_value)/1000}Bn', ha='center', va='bottom', color='navy', weight="bold", fontsize=15)
    ax2.text(i, max_value, f'Max: {int(max_value)/1000}Bn', ha='center', va='top', color='crimson', weight="bold", fontsize=15)


plt.tight_layout()
plt.show()
```

```python
# Insights into gender disparities in wealth accumulation
gender_wealth_stats = df.groupby('gender')['finalWorth'].describe()
print("\nStatistics of Wealth by Gender:")
display(gender_wealth_stats)

print("\nRichest Billionaire by Gender:")
display(df.loc[df.groupby('gender')['finalWorth'].idxmax()][["rank","finalWorth","category","personName","age","gender","country","source","industries"]])

gender_ratio = df['gender'].value_counts(normalize=True)
# Plot pie chart
plt.figure(figsize=(6, 4))
plt.pie(gender_ratio, labels=gender_ratio.index, autopct='%1.1f%%', colors=[palette[0], palette[4]])
plt.title('Gender Ratio Among Billionaires')
plt.show()
display(gender_ratio)
```

***The wealth acquired by billionaires in this list is divided among genders in the ratio of 87.2% for males and 12.8 for females. A total of 2303 of the billionaires are male and 337 are female.***

***Richest male is Bernard Arnault & family with a networth  of USD 211 Bn and richest female is Francoise Bettencourt Meyers & family with a networth of USD 80.50 Bn.***

### 1.3. Distribution of Wealth by Nationality

```python
# Set up the figure with subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 8))

# Distribution of Wealth by Nationality (Bar Plot)
sns.barplot(data=df, y='country', x='finalWorth', estimator=sum, ci=None, ax=ax1, order=df['country'].value_counts().index[:10], palette=palette)
ax1.set_title('Total Wealth by Nationality Among Billionaires (Top 10 Countries)', weight='bold',fontsize=12,pad=5)
ax1.set_xlabel('Total Wealth')
ax1.set_ylabel('Country')
for i in ax1.containers:
    ax1.bar_label(i, fmt='%d')

# Distribution of Wealth by Nationality (Box Plot)
sns.boxplot(data=df, y='country', x='finalWorth', ax=ax2, order=df['country'].value_counts().index[:10], color=palette[2])

ax2.set_title('Distribution of Wealth by Nationality Among Billionaires (Top 10 Countries)', weight='bold',fontsize=12,pad=5)
ax2.set_xlabel('Final Worth')
ax2.set_ylabel('Country')

# Add median and maximum labels
for i, country in enumerate(df['country'].unique()):
    median_value = df[df['country'] == country]['finalWorth'].median()
    max_value = df[df['country'] == country]['finalWorth'].max()
    if np.isfinite(median_value):
        ax.text(i, median_value, f'Median: {int(median_value)}', ha='center', va='bottom', color='black')
    if np.isfinite(max_value):
        ax.text(i, max_value, f'Max: {int(max_value)}', ha='center', va='bottom', color='red', weight='bold')


plt.tight_layout()
plt.show()
```

```python
# Set up the figure with subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 8))

# Distribution of Wealth by Nationality (Bar Plot)
sns.barplot(data=df, y='country', x='finalWorth', estimator=sum, ci=None, ax=ax1, order=df['country'].value_counts().index[:10], palette=palette)
for i in ax1.containers:
    ax1.bar_label(i, fmt='%d')
ax1.set_title('Total Wealth by Nationality Among Billionaires (Top 10 Countries)', weight='bold',fontsize=12,pad=5)
ax1.set_xlabel('Total Wealth')
ax1.set_ylabel('Country')

# Count of Billionaires per Country (Bar Plot)
country_counts = df['country'].value_counts().head(10)
sns.barplot(x=country_counts.values, y=country_counts.index, ax=ax2, palette=palette)
for i in ax2.containers:
    ax2.bar_label(i, fmt='%d')
ax2.set_title('Count of Billionaires by Nationality (Top 10 Countries)', weight='bold',fontsize=12,pad=5)
ax2.set_xlabel('Count')
ax2.set_ylabel('Country')

plt.tight_layout()
plt.show()
```

```python
# Calculate nationality ratio
nationality_ratio = df['country'].value_counts(normalize=True)

# Select top 5 countries and group others
top_countries = nationality_ratio.head(5)
other_countries_ratio = nationality_ratio[5:].sum()
top_countries['Others'] = other_countries_ratio

# Plot pie chart
palette = sns.color_palette("Blues_d", len(top_countries))
plt.figure(figsize=(6, 4))
plt.pie(top_countries, labels=top_countries.index, autopct='%1.1f%%', colors=palette)
plt.title('Nationality Ratio Among Billionaires (Top 5 Countries)', weight='bold',fontsize=12,pad=5)
plt.show()
```

***United state has the maximum wealth accumulated by billionaires with USD 4575.1Bn, followed by China with USD 1805.5Bn and India with USD 628.2Bn. United States has the highest number of billionaires with 754 billionaires in the country, followed by China with 523 and India at 157.***

### 2. Source of Wealth Distribution

### 2.1 Category head-count

```python
# Aggregate data by category and compute the count of billionaires
category_count = df['category'].value_counts().sort_values(ascending=False)

# Set up the figure with subplots
fig, axes = plt.subplots(1, 2, figsize=(18, 8))

# Plot count of billionaires by category (Bar Plot)
sns.barplot(x=category_count.index, y=category_count.values, palette=palette, ax=axes[0])

# Add bar labels
for i, value in enumerate(category_count.values):
    axes[0].text(i, value, f'{int(value)}', ha='center', va='bottom')

axes[0].set_title('Count of Billionaires by Category', weight='bold',fontsize=20,pad=50)
axes[0].set_xlabel('Wealth Category')
axes[0].set_ylabel('Count')

# Rotate and place category labels on the bars
axes[0].set_xticklabels(category_count.index, rotation=90)

# Visualize the proportion of the count of billionaires by category (Pie Chart)
axes[1].pie(category_count, labels=category_count.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('Blues_d', len(category_count)))
axes[1].set_title('Proportion of Billionaires by Category', weight='bold',fontsize=20,pad=50)
axes[1].axis('equal')

plt.tight_layout()
plt.show()
```

### 2.2 Distribution of Wealth among categories

```python
# Aggregate data by category and compute cumulative final worth
category_cumulative = df.groupby('category')['finalWorth'].sum().sort_values(ascending=False).cumsum()

# Set up the figure with subplots
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Plot cumulative final worth distribution by category (Bar Plot)
sns.barplot(x=category_cumulative.index, y=category_cumulative.values, palette=palette, ax=axes[0])

# Add bar labels
for i, value in enumerate(category_cumulative.values):
    axes[0].text(i, value, f'{int(value)/1000}', ha='center', va='bottom', rotation=45)

axes[0].set_title('Cumulative Final Worth Distribution by Category Among Billionaires', weight='bold',fontsize=12,pad=20)
axes[0].set_xlabel('Wealth Category')
axes[0].set_ylabel('Cumulative Final Worth in Bn USD')
# Rotate and place category labels on the bars
axes[0].set_xticklabels(category_count.index, rotation=90)

# Visualize the proportion of cumulative final worth among billionaires by category (Pie Chart)
axes[1].pie(category_cumulative, labels=category_cumulative.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('Blues_d', len(category_cumulative)))
axes[1].set_title('Proportion of Cumulative Final Worth Among Billionaires by Category', weight='bold',fontsize=12,pad=10)
axes[1].axis('equal')

plt.tight_layout()
plt.show()
```

### 2.3 Analysis of Self made wealth

```python
# Count the number of self-made and non-self-made billionaires
self_made_count = (df['selfMade'] == True).sum()
not_self_made_count = (df['selfMade'] == False).sum()

# Calculate the ratios
total_billionaires = len(df)
self_made_ratio = self_made_count / total_billionaires
not_self_made_ratio = not_self_made_count / total_billionaires

# Set up the figure with subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Plot count of self-made vs. inherited billionaires (Bar Plot)
sns.barplot(x=['Self-Made', 'Inherited'], y=[self_made_count, not_self_made_count], ax=axes[0], palette=palette)
axes[0].set_title('Count of Self-Made vs. Inherited Billionaires')
axes[0].set_ylabel('Count')

# Add bar labels
for i, value in enumerate([self_made_count, not_self_made_count]):
    axes[0].text(i, value, f'{value}', ha='center', va='bottom')


# Visualize the proportion of self-made vs. inherited billionaires (Pie Chart)
axes[1].pie([self_made_ratio, not_self_made_ratio], labels=['Self-Made', 'Inherited'], autopct='%1.1f%%', startangle=140, colors=[palette[0], palette[1]])
axes[1].set_title('Proportion of Self-Made vs. Inherited Billionaires')

plt.tight_layout()
plt.show()

# Display the ratios
print("Ratio of Self-Made Billionaires:", self_made_ratio)
print("Ratio of Inherited Billionaires:", not_self_made_ratio)
```

***Most number of billionaires have acquired their wealth from Finance & Investments followed by Manufacturing and Technology. A total of 372 (14.1%) billionaires acquired significant wealth from Finance & Investment, 324 (12.3%) from Manufacturing, and 214 (11.9%) from Technlogy***

***A total of 1812 billionaires accounting for 68.6% of the total number are self-made while 828 billionaires (31.4%) have acquired the fortune***

### 3. Top Industries of Billionaires

### 3.1 Identification of Predominant Industries

```python
# Aggregate data by industry and compute the count of billionaires in each industry
industry_counts = df['industries'].value_counts().head(10)

# Display the count of billionaires in each industry
print("Count of Billionaires by Industry:")
industry_counts
```

```python
# Set up the figure with subplots
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Calculate the cumulative final worth of the top 10 industries
industry_counts = df['industries'].value_counts().head(10)

# Visualize the cumulative final worth of the top 10 industries using a bar graph
sns.barplot(x=industry_counts.values, y=industry_counts.index, palette=palette, ax=axes[0])
axes[0].set_title('Count of Billionaires by Industry (Top 10)')
axes[0].set_xlabel('Count of Billionaires')
axes[0].set_ylabel('Industry')
axes[0].tick_params(axis='x', rotation=90)
  
# Add bar labels
for i, count in enumerate(industry_counts.values):
    axes[0].text(count, i, f'{count}', ha='left', va='center', fontsize=8, color='black')

# Calculate the count of billionaires in each industry to determine the prevalence
industry_counts = df['industries'].value_counts().head(10)

# Visualize the ratio of billionaires in each industry using a pie chart
axes[1].pie(industry_counts, labels=industry_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('Blues_d', len(industry_counts)))
axes[1].set_title('Ratio of Billionaires by Industry')
axes[1].axis('equal')

plt.tight_layout()
plt.show()
```

### 3.2 Distribution Across Industries

```python
# Set up the figure with subplots
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Calculate the cumulative final worth of the top 10 industries
cumulative_final_worth = df.groupby('industries')['finalWorth'].sum().sort_values(ascending=False).head(10)

# Plot distribution of cumulative wealth among industries using a KDE plot
sns.kdeplot(data=df, x='finalWorth', hue='industries', ax=axes[0], fill=True, palette=palette)
axes[0].set_title('Distribution of Cumulative Wealth Among Industries (KDE)')
axes[0].set_xlabel('Cumulative Final Worth')
axes[0].set_ylabel('Density')
axes[0].set_xlim(0, 50000)  # Set x-axis limits

# Plot distribution of cumulative wealth among industries using a box plot
sns.boxplot(data=df, x='finalWorth', y='industries', ax=axes[1], palette=palette)
axes[1].set_title('Distribution of Cumulative Wealth Among Industries (Box Plot)')
axes[1].set_xlabel('Cumulative Final Worth')
axes[1].set_ylabel('Industry')
    
plt.tight_layout()
plt.show()

display(cumulative_final_worth)
```

### 3.3 Geographic Variations in Industry Concentration

```python
# Aggregate Data by Country and Industry
country_industry_counts = df.groupby(['country', 'industries']).size().unstack(fill_value=0)

# 1. Aggregate Data by Country and Industry
country_industry_counts = df.groupby(['country', 'industries']).size().unstack(fill_value=0)

# 2. Select Top 10 Industries by Count of Billionaires
top_industries = df['industries'].value_counts().head(10).index
top_industry_counts = country_industry_counts[top_industries]
top_10_countries = top_industry_counts.sum(axis=1).nlargest(10).index
top_industry_counts_filtered = top_industry_counts.loc[top_10_countries]

# Create subplot for visualization
fig, axes = plt.subplots(1, 1, figsize=(10, 8))

# Visualize Top 10 Industries
sns.heatmap(top_industry_counts_filtered, cmap='Blues', annot=True, fmt='d', linewidths=.5, ax=axes)
axes.set_title('Top 10 Industries by Billionaire Count in Top 10 Countries')
axes.set_xlabel('Industry')
axes.set_ylabel('Country')

plt.tight_layout()
plt.show()
```

***Most number of billionaires have acquired their wealth from Finance & Investments (372), Manufacturing (324) and Technology (314)***

***In the US, as high as 190 billioniares hae reached the top spot are from Finance & Investment, followed by Technology (141)***

***In India, 32 billionaires hail from Manufacturing, 29 from Healthcare and 22 from Diversified***
